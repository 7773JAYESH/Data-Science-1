```python
from __future__ import print_function
import pandas as pd
import numpy as np
import folium

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
conf_df = pd.read_excel('ExcelData11.xlsx')
popul_df = pd.read_excel('ExcelData33.xlsx')
```

```python
world_map = folium.Map(location=[11,0], titles="cartodbpositron", zoom_start=2,max_zoom=6,min_zoom=2)

for i in range(len(popul_df)):
    folium.Circle(
    location=[popul_df.iloc[i]['Lat'],popul_df.iloc[i]['Long']],
    fill = True,
    radius = (int((np.log(popul_df.iloc[i,-1]+1.00001))) + 0.2)*50000,
    fill_color = 'blue',
    color = 'red',
    ).add_to(world_map)

world_map
```
