```python
import pandas as pd
import openpyxl
from openpyxl import load_workbook
import matplotlib.pyplot as plt

xl = pd.ExcelFile("ExcelData11.xlsx")
train = pd.read_excel('ExcelData11.xlsx')  #Reads data from Excel File
df = xl.parse("Sheet1")  #Creates DataFrame in Python

a = df['Overseas Indian']  #Stores value of column 'Overseas Indian'
b = df['Persons of Indian Origin (PIOs)']  #Stores value of 'Persons of Indian Origin (PIOs)' column in "ExcelData11.xlsx" file
b,a
plt.plot(a,b)  #Plots Graph in Cartesian Form
plt.ylabel('Persons of Indian Origin')    #Writes axis text in y-axis(adds label to specified target object)
plt.xlabel('Overseas Indians')            #Writes axis text in x-axis(adds label to specified target object)
plt.title('Migrants Charts')              #Writes text on top of the plot (adds label on top of object)

#Conclusion: Analysed dataset using line graph
```

```python
plt.plot(a,b,'k-',a,b,'r^')
plt.xlabel('Persons of Indian Origin (PIOs)')
plt.ylabel('Overseas Indians')

#Conclusion: Learned to analyse data using Line & markers using python,matplotlib.Plotly
```

```python
name = ['Overseas Indian','Persons of Indian Origin (PIOs)']
value = ['a','b']

plt.bar(name,value)   #Plots a Bar Graph
plt.title('Bar Ggraph of Migrants True/False')
plt.xlabel('Migrants')

#Conclusion: Analysed data using python & matplotlib.plotly
```
